```python
import geopy
geocoder=geopy.geocoders.GoogleV3(domain="maps.google.co.uk")
def geolocate(place):
  return geocoder.geocode(place,exactly_one=False)[0][1]
```

```python
london_location=geolocate("London")
print london_location
```

```python
from nose.tools import assert_almost_equal
assert_almost_equal(london_location[0],51.51,places=2)
assert_almost_equal(london_location[1], -0.125,places=2)
```

```python
def maps_url_for(lat,long, satellite=False,zoom=12,size=(400,400),sensor=False):
  url="http://maps.googleapis.com/maps/api/staticmap?"
  params=dict(
    sensor= str(sensor).lower(),
    zoom= zoom,
    size= "x".join(map(str,size)),
    center= ",".join(map(str,(lat,long))),
    style="feature:all|element:labels|visibility:off"
  )
  if satellite:
    params["maptype"]="satellite"

  paramstring="&".join([key+"="+str(value) for
    key,value in params.items()])

  return url+paramstring
```

```python
url=maps_url_for(51.5072, -0.1275, zoom=10)
print url
```

```python
from nose.tools import assert_in
assert_in("http://maps.googleapis.com/maps/api/staticmap?",url)
assert_in("center=51.5072,-0.1275",url)
assert_in("zoom=10",url)
assert_in("size=400x400",url)
assert_in("sensor=false",url)
```

```python
import png
import urllib2
def get_map_at(lat,long,**args):
  return urllib2.urlopen(maps_url_for(lat,long,**args))
```

```python
import IPython
map_png=get_map_at(*london_location)
IPython.core.display.Image(map_png.read())
```

```python
def is_green(r,g,b):
  threshold=1.1
  return g>r*threshold and g>b*threshold
```

```python
import png
from itertools import izip

def count_green_in_png(data):
    image=png.Reader(file=data).asRGB()
    count = 0
    for row in image[2]:
        pixels=izip(*[iter(row)]*3)
        count+=sum(1 for pixel in pixels if is_green(*pixel))
    return count
```

```python
print count_green_in_png(get_map_at(*london_location))
```

```python
from StringIO import StringIO
def show_green_in_png(data):
    image=png.Reader(file=data).asRGB()
    count = 0
    out=[]
    for row in image[2]:
        outrow=[]
        pixels=izip(*[iter(row)]*3)
        for pixel in pixels:
            outrow.append(0)
            if is_green(*pixel):
                outrow.append(255)
            else:
                outrow.append(0)
            outrow.append(0)
        out.append(outrow)
    buffer=StringIO()
    result = png.from_array(out,mode='RGB')
    result.save(buffer)
    return buffer.getvalue()
```

```python
IPython.core.display.Image(get_map_at(*london_location,satellite=True).read())
```

```python
IPython.core.display.Image(show_green_in_png(get_map_at(*london_location,satellite=True)))
```

```python
from numpy import linspace
def location_sequence(start,end,steps):
  # Would actually prefer this if steps
  # were deduced from zoomlevel
  # But need projection code for that
  lats=linspace(start[0],end[0],steps)
  longs=linspace(start[1],end[1],steps)
  return zip(lats,longs)
```

```python
print location_sequence(geolocate("London"),geolocate("Cambridge"),10)
```

```python
for location in location_sequence(geolocate("London"),geolocate("Birmingham"),10):
    IPython.core.display.display(IPython.core.display.Image(get_map_at(*location,zoom=10,satellite=True).read()))
```
